### **MATPLOTLIB EXAMPLE WORKFLOW**

```python
#0 - import matplotlib and get it ready for plotting i
import matplotlib.pyplot as plt

#1 - prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2 - setup plot
fig, ax = plt.subplots()

#3 - plot data
ax.plot(x,y)

#4 - customize plot
ax.set(title='Simple Plot', xlabel='x', ylabel='y')

#5 - save & show plot
fig.savefig('simple_plot.png')
plt.show()
```

### TWO OPTIONS FOR SUBPLOTS

```python
#Subplot option   1
fig, ((ax1, ax2),(ax3,ax4)) =  plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10),np.random.randn(10));
ax3.bar(x, x/2);
ax4.hist(np.random.randn(1000))
```

```python
#subplots options 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.randn(10),np.random.randn(10));
ax[1,0].bar(x, x/2);
ax[1,1].hist(np.random.randn(1000))
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True) # Tell pandas to replace using regex
```

```python
car_sales
```

```python
#remove last 2 zeroes

car_sales['Price'] = car_sales['Price'].str[:-2]
```

```python
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2025', periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"] = car_sales['Price'].cumsum() #this is in str therefore the result came out in addition of the number as it would with an integer
car_sales
```

```python
car_sales["Total sales"] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#plotting the total sales
car_sales.plot(x="Sale Date", y="Total sales")
```

```python
car_sales["Price"] = car_sales["Price"].astype(str)

# Plot a scatter plot
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
