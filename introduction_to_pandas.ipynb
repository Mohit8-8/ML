```python
#column from series
seats_column =  pd.Series([5,5,5,5,5])

#new column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales['Seats'].fillna(car_sales['Seats'].mean(),inplace=True)
car_sales
```

```python
#create a column from list
#it has to have exactly same number of index as the original indexes unlike the series there can be no empty space
fuel_economy=[7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales['Fuel per 100KM'] = fuel_economy
car_sales
```

```python
car_sales['Total Fuel Used (L)'] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100KM']
```
