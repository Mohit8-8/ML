### **NOW WE ARE GOING TO STANDARDIZE THE DATA **
Standardization is a data preprocessing technique used in machine learning to transform numerical features so that they have zero mean and unit variance.

by centering the data around zero and scaling it to have a standard deviation of one.

x_standardized = (x - mean(x)) / std(x)
